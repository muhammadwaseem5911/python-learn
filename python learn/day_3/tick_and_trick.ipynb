# splitting into string to multiple column

```python
#import libraries
import  pandas as pd
df=pd.DataFrame({'name':['umair afzal','saad afzal','jawad afzal','mahnoor akhtar'],
                'location': ['ghura,pak','sara,pak','karchai,pak','bulani,pak']})
df
```

```python
#splitting

df[['first_name','second_name']]=df.name.str.split(' ', expand=True)
```

```python
df
```

```python
#split loction into two column 
#and also add new column 
#when you adding column must use [[]] these brackets
df[['village','country']]=df.location.str.split(',',expand=True)
```

```python
df
```
